```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names() # 연습 데이터 list
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# sns.countplot(x='class', data=titanic) # 빈도수
sns.countplot(titanic['class'])
```

```python
sns.countplot(x='class', hue='sex', data=titanic) # 서브 데이터
```

```python
sns.barplot(x='class', y='age', data=titanic)
```

```python
age = titanic.age.values
sns.rugplot(age)
```

```python
sns.displot
```

```python
sns.kdeplot(age) # 분포곡선
```

```python
sns.displot(age, kde=True, rug=True) # 한번에 그릴수있음
```

```python
sns.jointplot( x='age', y='fare', data=titanic, kind='kde')
```

```python
sns.pairplot(titanic)
```

```python
# 10주차 2차시
titanic_2 = titanic.pivot_table(index='class',columns='sex',aggfunc='size')
sns.heatmap(titanic_2,annot=True,fmt='d',cmap= sns.light_palette('red',as_cmap= True))
```

```python
# aggfunc = 값
# columns = x index = y annot = 수치값, fmt = 타입 d, .f .2f 
# cmap = 컬러맵 as_cmap

# 복합 데이터 그래프 그리기 분포를 쉽게 눈으로 확인할 수 있게해줌
sns.violinplot(x='embark_town', y='age', data=titanic)
```

```python
sns.stripplot(x='embark_town', y='age',data=titanic)
```

```python
sns.swarmplot(x='embark_town', y='age',data=titanic)
```

```python
sns.boxplot(x='embark_town', y='age',data=titanic)
# 데이터 사분위값
```
