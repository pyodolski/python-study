```python
 casts = pd.read_csv('cast.csv')
```

```python
casts.fillna(0, inplace=True)
```

```python
# 8번 casts에서 'Abrams'가 출연한 영화를 모두 찾아서 cf에 저장
cf = casts[casts.name == 'Aaron Abrams']
```

```python
cf
```

```python
cf.title.value_counts()
```

```python
# 연도별로 출연한 영화의 수 막대 그래프
cf2 = cf.groupby('year').size()
```

```python
cf2
```

```python
import matplotlib.pyplot as plt
```

```python
cf2.plot(kind='bar')
plt.show()
```

```python
cf2.plot(kind='barh')
plt.show()
```

```python
# 10번 casts에서 각 각 decade 마다 몇 편의 영화가 나왔는지 구하시오
decades = casts['year'] // 10 * 10 
```

```python
df2 = casts.groupby(decades).size()
```

```python
df2.plot(kind='bar')
plt.show()
```

```python
casts.head(2)
```

```python
df3 = casts.groupby([decades, 'type']).size()
```

```python
df3.plot(kind='bar')
```

```python
df4 = casts.groupby(['type', decades]).size()
```

```python
df4
```

```python
df4.plot(kind='bar')
```

```python
#multindex -> colum : unstack
df4.unstack()
```
