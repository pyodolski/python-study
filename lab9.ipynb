```python
import matplotlib.pyplot as plt
```

```python
x1 = [10,20,30]
y1 = [20,40,10]
x2 = [10,20,30]
y2 = [40,10,30]
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x1,y1,color='blue',linewidth=3,label='dotted', linestyle='dotted')
plt.plot(x2,y2,color='red',linewidth=3,label='dashed', linestyle='dashed')
plt.title("YU CSE")
plt.legend() # 레이블 정보 표시
plt.show()
```
